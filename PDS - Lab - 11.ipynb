# WAP to implement linear regression with iphone csv

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("iphone.csv")
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
m = LinearRegression()
```

```python
x = df['number']
y = df['Price']
```

```python
newx = x.values.reshape(-1,1)
newx
```

```python
m.fit(newx,y)
```

```python
predictval = m.predict(newx)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.scatter(x,predictval)
```
