```python
x = range(1,11,1)
cxMarks = [8,9,6,3,5,7,4,1,2,6]
cyMarks = [5,8,9,6,3,2,4,8,8,9]
plt.plot(x,cxMarks,marker=">",ls="--",color="red")
plt.plot(x,cyMarks,marker="o",color="blue")
plt.xlabel("---------- RoII No ------------")
plt.ylabel("Cpi")

plt.annotate(xy=[8,1],text='Lowest Cpi')
plt.legend(["CX","CY"])
```

# Generate scatter plot grouped by smoker field and 
# age on x axis and charges on y axis
df = pd.read_csv('insurance.csv')

```python
df = pd.read_csv('insurance.csv')

grouped = df.groupby(['smoker'])
for key,group in grouped:
    plt.scatter(group['age'],group['charges'], label=key)

plt.legend(['Smoke : no','Smoke : yes'])
plt.show()
```
