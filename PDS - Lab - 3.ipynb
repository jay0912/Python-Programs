# 01) WAP to check given string is palindrome or not.

```python
str = input("Enter String to check Palindrom : ")
reverse = str[::-1]

if str==reverse:
    print("Given String ",str," is Palindrom.")
else :  
    print("Given String ",str," is not Palindrom.")
```

# 06) WAP to count numbers of vowels in given string.

```python
str = input("Enter String : ")
count = 0

vowels = ('a','e','i','o','u')

for i in str.lower():
    if i in vowels :
        count = count + 1

print("Total Vowels in the Given String : ",count)        
    
```

# 08) WAP to check whether string contains any special character or not.

```python
str = input("Enter String : ")

Special_Ch = ('[@_!#$%^&*()<>?/\|}{~:]')

flag = False

for i in str:
    if i in Special_Ch:
        flag = True
        break

if flag:
    print("Contains Special Character : ",i)
else :
    print("Contains No Special Characters.")
```

# 09) WAP to generate random strings until a given string is generated.

```python
import random
import string

count=0
result=""

str = input("Enter the strings:")
print("input Stirng : ",str)

while str!=result:
    result = "".join(random.choice(string.ascii_letters) for i in range(len(str)))
    print(result)
    count += 1

print("Number of randoms generated:",count)
```

# 10) WAP to check given string data in binary formats or not. 

```python
str = input("Enter String : ")
print("Input String : ",str)

Binary = ('0','1')

flag = False

for i in str:
    if i in Binary:
        flag = True
    else :  
        flag = False
        break

if flag:
    print("Given String is in Binary format")
else :
    print("Given String is not in Binary format")
```

# 11) WAP to find all duplicate characters in string.

```python
s = input("Enter a string : ")
duplicates = {}

for i in s:
  if i in duplicates:
    duplicates[i] += 1
  else:
    duplicates[i]=1
    
for i,j in duplicates.items():
  if(j>1):
    print(i," ")
```

# 13) WAP to do String slicing
                 a)	Left (Or anticlockwise) rotate the given string by d elements (where d <= n).
                 b)	Right (Or clockwise) rotate the given string by d elements (where d <= n).


```python
s=input("Enter a string : ")
index = int(input("Which element you want to change : "))
firstLeft = s[index:]
lastLeft = s[0:index]
firstRight = s[0:len(s)-index]
lastRight = s[len(s)-index:]

print("Left rotation : ", firstLeft+lastLeft)
print("Right rotation : ",lastRight +firstRight)
```
