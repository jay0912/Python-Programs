# 1)	Scrape the data of Faculty and news data from darshan university website.



```python
import bs4
import requests

req = requests.get("https://www.darshan.ac.in/engineering/computer/faculty")
soup = bs4.BeautifulSoup(req.text,"lxml")
allFaculty = soup.select("body > main > div:nth-child(5) > div > div > div.col-lg-8.col-xl-9.g-font-size-14 > div > div ")
print("Faculty List:")
for fac in allFaculty:
     allSpans = fac.select("h2")
     print("\t",allSpans[0].text.strip())
    
req = requests.get("https://www.darshan.ac.in/engineering/electrical/news")
soup = bs4.BeautifulSoup(req.text,"lxml")
news = soup.select("body > main > div:nth-child(5) > div > div > div.col-lg-8.col-xl-9.g-font-size-14.g-mb-30 > div > div")
print("\nNews:")
for part in news[1:]:
    h2 = part.select("h2")[0].text.strip()
    print("\t"+h2)
```

# 2) Scrape the data of student placement from darshan university website.

```python
import requests
import bs4

req = requests.get("https://www.darshan.ac.in/placement/list/engineering/computer/2023")
soup = bs4.BeautifulSoup(req.text,"lxml")
placements = soup.select("body > main > div:nth-child(5) > div > div > div.col-lg-8.col-xl-9.g-font-size-14.g-mb-30 > div.g-pt-20 > div.row.g-px-4.g-px-5--sm.g-px-5--md > div")
for placement in placements:
    item = placement.select("span")[0].text.strip()
    print(item)
```

```python
import requests
import bs4

req = requests.get("https://www.darshan.ac.in/placement/list/engineering/computer/2023")
soup = bs4.BeautifulSoup(req.text,"lxml")
placements = soup.select("body > main > div:nth-child(5) > div > div > div.col-lg-8.col-xl-9.g-font-size-14.g-mb-30 > div.g-pt-20 > div.row.g-px-4.g-px-5--sm.g-px-5--md > div")
for placement in placements:
    name = placement.select("span")[0].text.strip()
    package = placement.select("span")[1].text.strip()
    print(name,":",package)
```

# Write a python program to perform database operations on database using mysql and sqlite3.

```python
import sqlite3
connection = sqlite3.connect("abc.db")
crsr = connection.cursor()
print("Connected to the database")
#connection.close()

#create student table
sql_command = """CREATE TABLE Student_table (
roll_no INTEGER PRIMARY KEY,
fname VARCHAR(50),
lname VARCHAR(50),
cpi FLOAT,
city VARCHAR(20));"""
crsr.execute(sql_command)
#connection.close()

sql_command = """INSERT INTO Student_table VALUES 
(1, "abc", "xyz", 9.2, "Rajkot");"""
crsr.execute(sql_command)
sql_command = """INSERT INTO Student_table VALUES 
(2, "def", "pqr", 8.9, "Rajkot");"""
crsr.execute(sql_command)
sql_command = """INSERT INTO Student_table VALUES 
(3, "ghi", "stu", 7, "Jamnagar");"""
crsr.execute(sql_command)
sql_command = """INSERT INTO Student_table VALUES 
(4, "jkl", "sqr", 9.21, "Jamnagar");"""
crsr.execute(sql_command)
sql_command = """INSERT INTO Student_table VALUES 
(5, "mno", "mno", 8.66, "Morbi");"""
crsr.execute(sql_command)
connection.commit()
#connection.close()

crsr.execute("SELECT * FROM Student_table;")
ans = crsr.fetchall()
for i in ans:
  print(i)



rno = input("Enter roll number : ")
f_nm = input("Enter first name : ")
l_nm = input("Enter last name : ")
cpi = input("Enter CPI : ")
ct = input("Enter city : ")
 
crsr.execute("""
INSERT INTO Student_table(roll_no, fname, lname, cpi, city)
VALUES (?,?,?,?,?)
""", (rno, f_nm, l_nm, cpi, ct))

connection.commit()
#connection.close()


crsr.execute("SELECT * FROM Student_table;")
ans = crsr.fetchall()
for i in ans:
  print(i)
```
